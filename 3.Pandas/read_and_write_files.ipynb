## Concatenate Data Frames

```python
import pandas as pd
india_weather=pd.DataFrame({
    "city":['Kolkata','Bangalore','Mumbai'],
    "temperature":[30,32,40],
    "humidity":[80,60,78]
})
india_weather
```

```python
import pandas as pd
us_weather=pd.DataFrame({
    "city":['newyork','chicago','orlando'],
    "temperature":[21,14,35],
    "humidity":[52,81,78]
})
us_weather
```

```python
#concatenate datframes
pd.concat([india_weather,us_weather])
```

```python
#to ignore the index
pd.concat([india_weather,us_weather],ignore_index=True)
```

```python
# row wise concatenation
pd.concat([india_weather,us_weather],axis=1)
```

# Merging Dataframes

```python
import pandas as pd
temperature_df=pd.DataFrame({
    "city":['Kolkata','Bangalore','Mumbai','Hyderabad'],
    "temperature":[30,32,40,28],
    "humidity":[80,60,78,67]
})
temperature_df
```

```python
humidity_df=pd.DataFrame({
    "city":['Kolkata','Bangalore','Mumbai'],
    "temperature":[30,32,40],
    "humidity":[80,60,78]
})
humidity_df
```

```python
# merging the two dataframe based on city
pd.merge(temperature_df,humidity_df,on='city')
```

```python
#to merge the hyderabad column also
pd.merge(temperature_df,humidity_df,on='city',how='outer')
```
