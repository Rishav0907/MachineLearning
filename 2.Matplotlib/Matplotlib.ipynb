# Matplotlib Pyplot

#### Based on the famous MATLAB

```python
from matplotlib import pyplot as plt
plt.plot([2,4,6,4])
plt.xlabel("Number")
plt.ylabel("indices")
plt.title("My plot")
plt.show()
# only y axis values are given. So the index values becomes x values
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('square')
plt.ylabel('numbers')
plt.title('grid box')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') # r means red colour and o means o shaped dots
plt.grid()
plt.show()
```

```python
import numpy as np
t=np.arange(0,5.,0.2)

plt.plot(t,t**2,'b--',label='^2') # b-- means blue dashed line
plt.plot(t,t**2.2,'rs',label='^2.2') #rs means red coloured squares
plt.plot(t,t**2.5,'g^',label='^2.5') #g^ means traingle
plt.legend() #legend is the scale in the graph
plt.grid()
plt.show()
```

## Controlling the properties

```python
x=[5,9,2,3]
y=[7,1,8,6]
plt.plot(x,y,linewidth=5)# to increase width of line
plt.show()
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
line=plt.plot(x1,y1,x2,y2)
plt.setp(line[0],color='red',linewidth=2)
plt.setp(line[1],color='blue',linewidth=2)
plt.grid()
plt.show()
```

## Working with multiple figure and axis

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(1)#creating a figure for me

plt.subplot(211)# the subplot function shows no. of subplots,no of columns and the figure no
plt.grid()
plt.plot(t1,f(t1),'b--')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r-')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3],'b-')
plt.subplot(212)
plt.plot([4,5,6],'r--')

plt.figure(2)
plt.plot([4,5,6],'ro')

plt.figure(1)
plt.subplot(211)
plt.title("Easy")
plt.show()
```
